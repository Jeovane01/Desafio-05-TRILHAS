```python
#Importando Bibliotecas
import pandas as pd
import seaborn as sns
```

```python
#Importando DataFrame
dados = pd.read_csv('/content/Planilha do desafio 5 - Ciência de dados - Relatório do Banco x.csv')
```

```python
#Visualização preliminar dos dados
dados.head()
```

```python
#Visualizando as colunas do DataFrame
dados.columns
```

```python
#Substituir o cabeçalho por palavras em letras maiúsculas;
dados = dados.rename(columns={'Id': 'ID', 'Pontos': 'PONTOS', 'Estado': 'ESTADOS', 'Genero': 'GENERO', 'Idade': 'IDADE',
                              'Bens': 'BENS', 'Saldo na Conta': 'SALDO NA CONTA', 'Produto': 'PRODUTO', 'Possui cartao':
                              'POSSUI CARTAO', 'Ativo': 'ATIVO', 'Salario': 'SALARIO', 'Saiu':'SAIU'})
dados
```

```python
#Corrigir os dados categóricos, por exemplo, "Mas" para "Masculino" e "Fem" para "Feminino" para adequá-los ao padrão dos demais;
dados['GENERO'].replace(['M', 'F', 'Fem'], ['Masculino', 'Feminino', 'Feminino'], inplace = True)
dados
```

```python
#Visualizando O DataFrame filtrando o gênero Feminino
dados.loc[dados['GENERO']== 'Feminino']
```

```python
#Removendo valores 'nan' da coluna GENERO
dados = dados[dados['GENERO'].notnull()]
dados
```

```python
#Visualizando Idades presentes no arquivo
dados['IDADE'].unique()
```

```python
#Removendo Outlines
#Preenchendo com a média das idades
media_idades = dados.IDADE.mean()
dados.fillna(media_idades, inplace= True)
```

```python
#Corrigindo os dados faltantes e preencher as lacunas com valores adequados
#Preenchendo os valores com a média dos salários
media_salario = dados.SALARIO.mean()
dados.fillna(media_salario, inplace= True)
dados.SALARIO.isnull().sum()
```

```python
#Distribuição das Idades
Dist_Idades = sns.distplot(dados.IDADE)
Dist_Idades.figure.set_size_inches(12, 6)
Dist_Idades
```

```python
#Distribuição das Idades
Dist_Saldo_Na_Conta = sns.distplot(dados["SALDO NA CONTA"])
Dist_Saldo_Na_Conta.figure.set_size_inches(12, 6)
Dist_Saldo_Na_Conta
```

```python
#De acordo com o histograma de saldo na conta infere-se que há muitas pessoas sem saldos em suas conta!
```

```python
# Média, Mediana e Moda das idades das pessoas
media = dados.IDADE.mean()
mediana = dados.IDADE.median()
moda = dados.IDADE.mode()
```

```python
print('Média:',media.round(),'Mediana:', mediana,'Moda:', moda[0])
```

```python
#Quantidade de Homens e Mulheres
frequencia = dados['GENERO'].value_counts()
frequencia
```

```python
#Porcentagem de Homens e Mulheres
frequencia_percentual = dados['GENERO'].value_counts(normalize = True)*100
frequencia_percentual
```

```python
# BoxPlot da quantidade de pontos acordo com o gênero
salario_por_genero = sns.boxplot(x = 'PONTOS', y = 'GENERO', data = dados, orient = 'h')
salario_por_genero.figure.set_size_inches(12,4)
salario_por_genero.set_title('Pontos Por Gênero', fontsize=18)
salario_por_genero.set_xlabel('Pontos', fontsize=14)
salario_por_genero
```

```python
# Os homens possuim um pouco a mais de ponto comparada às mulheres.
```

```python
# BoxPlot da quantidade de bens de acordo com o gênero
bens_por_genero = sns.boxplot(x = 'BENS', y = 'GENERO', data = dados, orient = 'h')
bens_por_genero.figure.set_size_inches(12,4)
bens_por_genero.set_title('Quantidade de Bens Por Gênero', fontsize=18)
bens_por_genero.set_xlabel('Quantidade de Bens', fontsize=14)
bens_por_genero
```

```python
# As mulheres possuem menos bens quando se compara com os bens dos homens.
```
